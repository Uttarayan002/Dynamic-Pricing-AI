# CUSTOMER FEEDBACK SIGNALS
## Goal: Does better customer rating drive sales?
> Reasoning: Understand whether ratings significantly influence repeat purchases or price justification.


```python
import pandas as pd

data = pd.read_csv('finalweekly_modeldataset_cleaned.csv')
df = data.copy()
```

```python
# Rating vs Weekly Order Quantity

import plotly.express as px
fig = px.scatter(df, x='avg_rating', y='weekly_order_qty', color='category',
                 size='num_reviews', title='Rating vs Weekly Orders (Bubble = Reviews)')
fig.show()
```

# Insights
Customer Ratings vs Weekly Orders (Bubble Size = Reviews)
Products with high average ratings (4.0–5.0) do not always receive the highest weekly orders, indicating that customer ratings alone are not strong predictors of sales volume.
However, products with a higher number of reviews tend to correlate with higher weekly order quantities, suggesting that social proof (engagement volume) plays a bigger role in influencing purchase behavior than rating quality alone.
Additionally, category-level differences imply that product-type influences how much ratings impact sales.
